## 값을 여러 개 출력하기

```python
print(1,2,3)
```

```python
print('Hello', 'Python')
```

### sep로 값 사이에 문자 넣기

```python
print(1,2,3, sep=', ')
```

```python
print('Hello', 'Python', sep=', ')
```

```python
print(1920,1080, sep=' X ')
```

## 줄바꿈 활용

```python
print(1, 2, 3, sep='\n')
```

```python
print('1\n2\n3')
```

### end 사용

```python
print(1,end='')
print(2,end='')
print(3)
```

```python
print(1,end=' ')
print(2,end=' ')
print(3)
```

## 연습문제

```python
year = 2000
month = 10
day = 27
hour = 11
minute = 43
second = 59
```

```python
print(year,month,day,sep='/',end=' ')
print(hour,minute,second,sep=':')
```

```python
year, month, day, hour, minute, second = input().split()
print(year,month,day,sep='-',end=' T ')
print(hour,minute,second,sep=':')
```
